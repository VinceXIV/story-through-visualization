# Story Telling Using Data

I will be doing story telling using exchange rates data available from [kaggle](https://www.kaggle.com/lsind18/euro-exchange-daily-rates-19992020) 

## 1. Initial Set Up and Data Exploration

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime

%matplotlib inline
```

```python
exchange_rates = pd.read_csv('euro-daily-hist_1999_2020.csv')
```

## 2. Data Cleaning

### 2.1 Rename "US dollar" and "Period\\Unit" Columns

```python
exchange_rates.rename(columns={'[US dollar ]': 'US_dollar',
                               'Period\\Unit:': 'Time'},
                      inplace=True)
exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])
exchange_rates.sort_values('Time', inplace=True)
exchange_rates.reset_index(drop=True, inplace=True)
```

```python
euro_to_dollar = exchange_rates.loc[:, ['Time', 'US_dollar']]
```

### 2.2 Drop Records with "-" as Values for US_dollar

Some records (62) have the value for US_dollar variable set to "-". This is weird, I will be dropping it

```python
euro_to_dollar['US_dollar'].value_counts()
```

```python
euro_to_dollar = euro_to_dollar[euro_to_dollar['US_dollar'] != "-"]
```

### 2.3 Convert US_dollar to type float

```python
euro_to_dollar['US_dollar'] = euro_to_dollar['US_dollar'].astype(float)
```

```python
euro_to_dollar.info()
```

## 3. Story through Plot

```python
import matplotlib.style as style
style.use('ggplot')
```

```python
euro_to_dollar['rolling_mean'] = euro_to_dollar['US_dollar'].rolling(30).mean()
```

```python
euro_to_dollar['year'] = euro_to_dollar['Time'].dt.year
```

```python
year_btwn_2006_and_2010 = (euro_to_dollar['year'] >= 2006) & (euro_to_dollar['year'] <= 2009)
euro_to_dollar = euro_to_dollar[year_btwn_2006_and_2010]
grey_area = ((euro_to_dollar['Time'].dt.year > 2008)
             & (euro_to_dollar['Time'].dt.month > 2)
             & (euro_to_dollar['Time'].dt.year < 2009)
             & (euro_to_dollar['Time'].dt.month > 6))

fig, ax = plt.subplots(figsize=(10, 4), facecolor=(0.8980392156862745, 0.8980392156862745, 0.8980392156862745, 1.0))
ax.plot(euro_to_dollar['Time'],
         euro_to_dollar['rolling_mean'],
         color='green',
         linewidth=1.2,
         alpha=0.3)

year_btwn_2007_and_2009 = (euro_to_dollar['year'] >= 2007) & (euro_to_dollar['year'] <= 2008)

ax.plot(euro_to_dollar.loc[year_btwn_2007_and_2009, 'Time'],
         euro_to_dollar.loc[year_btwn_2007_and_2009, 'rolling_mean'],
         color='red',
         linewidth=2.5)
ax.plot(euro_to_dollar.loc[year_btwn_2007_and_2009, 'Time'],
         euro_to_dollar.loc[year_btwn_2007_and_2009, 'rolling_mean'],
         color='red',
         linewidth=2.5)

curr_xticks = ax.get_xticks()
ax.axvline(x=curr_xticks[5]-20,
           linewidth=80,
           color='grey',
           alpha=0.3,
           ymin=0.26)

ax.set_xticklabels([2006, "", 2007, "", 2008,"", 2009,"", 2010])
ax.tick_params(bottom=False, left=False)
ax.set_xlim([datetime(2005, 3, 1), datetime(2010, 1, 1)])
ax.grid(alpha=0.5, color='grey')
ax.text(x=curr_xticks[0]-80,
        y=1.67,
        s="Euro-USD rate peaked at 1.59 during 2007-2008's financial crisis",
        weight='bold', size=15)
ax.text(x=curr_xticks[0]-80,
        y=1.63,
        s="Euro-USD exchange rate between 2006 and 2010",
        size=15)
ax.set_xlim(datetime(2006, 1, 1), datetime(2010, 1, 1))
ax.set_yticks([1.2, 1.3, 1.4, 1.5])
positions = ['left', 'right', 'bottom', 'top']
for position in positions:
    ax.spines[position].set_visible(False)
    
ax.text(x=curr_xticks[0]-80,
        y=1.1,
        s='Â©Vincent Makokha'+' '*110+ 'Source: European Central Bank',
        backgroundcolor='#4d4d4d',
        color='white')
```
